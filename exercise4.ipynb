```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

from pandas import Series, DataFrame

np.set_printoptions(precision=2, suppress=True, linewidth=120)

pd.options.display.max_rows = 20

plt.rc("figure", figsize=(8, 4))
```

## Data preparation

Read the data from the provided *fortune500.csv* into a `DataFrame`.

Drop "(in millions)" from the names of the columns so that dot-notation can be used to access them.

Set data type of the *Company* column to a string and data type of the "Profit (in millions)" to a floating-point number.

```python
df = pd.read_csv("fortune500.csv")

df.rename({"Revenue (in millions)": "Revenue", "Profit (in millions)": "Profit"},
          axis="columns",
          inplace=True)

df.Company = df.Company.astype(pd.StringDtype())
df.Profit = pd.to_numeric(df.Profit, errors="coerce")
```

## Task 4

Plot IBM's (*Intl. Business Machines*) annual rank between 1965 and 2005, annotating the following dates:

- 1970: "Relational DB"
- 1981: "IBM PC"
- 1997: "Deep Blue"

The `y`-axis should be inverted and limited to `[0..15]` to make changes in ranking more visible.

Your figure should look similar (though not necessarily identical) to the following:

![Task 4](task4.png)

```python
# Add your implementation here.
IBM_data = df[df["Company"] == "Intl. Business Machines"]
```

```python
IBM_data
```

```python
IBM_data = IBM_data[IBM_data['Year'] > 1965]
ibm_rank_df = IBM_data[['Year', 'Rank']]

# Create the plot
plt.figure(figsize=(10,5))
plt.plot(ibm_rank_df['Year'], ibm_rank_df['Rank'])

# Add the annotations
plt.annotate('Relational DB', xy=(1970, 6), xytext=(1967.5, 10), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('IBM PC', xy=(1981, 9), xytext=(1979.75, 14), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('Deep Blue', xy=(1997, 7), xytext=(1995, 10), arrowprops=dict(facecolor='black', arrowstyle='->'))

# Set the y-axis limits and invert it
plt.ylim(15, 0)

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Rank')
plt.title('IBM')

# Display the plot
plt.show()
```
